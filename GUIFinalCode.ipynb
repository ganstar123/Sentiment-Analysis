```python
import pandas as pd
import torch
import numpy as np
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer
import warnings
import sqlite3
import tkinter as tk
from tkinter import *

#my_variable = 0

# Set pandas display options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
warnings.filterwarnings("ignore")

# Connect to SQLite database
# Read product data from SQLite
database_connection = sqlite3.connect("amazon_reviews_database.db")
# Read product data from a SQL database
product_data_query = "SELECT * FROM product_data"
product_data = pd.read_sql_query(product_data_query, database_connection)
database_connection.close()
print(product_data.head())
database_connection = sqlite3.connect("amazon_reviews_database.db")

# Read merged data from SQLite and concatenate additional tables
merged_data = pd.read_sql_query("SELECT * from ECMB000082", database_connection)
for i in range(2, len(product_data) + 1):
    merged_data_temp = pd.read_sql_query("SELECT * from ECMB{:06d}".format(i), database_connection)
    merged_data = pd.concat([merged_data, merged_data_temp])
# Close database connection    
database_connection.close()
print(merged_data.head())
import missingno as msno

# Visualize missing data using missingno
plt.figure(figsize=(25, 20))
msno.matrix(merged_data, color=[0, 0, 0])
plt.show()

# Drop rows with missing values
merged_data.dropna(inplace=True, axis=0)
print(merged_data.info())
import missingno as msno
plt.figure(figsize=(25, 20))
msno.matrix(merged_data, color=[0, 0, 0])
plt.show()
import re

# Define function to remove URLs from text
def remove_url(text):
    text = re.sub(r"http\S+", "", text)
    return text

# Install and import necessary libraries
!pip install demoji
import demoji
# Function to handle emojis in text
def handle_emoji(string):
    emojis = demoji.findall(string)
    for emoji in emojis:
        string = string.replace(emoji, " " + emojis[emoji].split(":")[0])
    return string
# Install spellchecker library
from spellchecker import SpellChecker
spell_data = SpellChecker()
# Add misspelled words to the list
misspelled_words = ["nsme", "mobiile", "phoone"]
# Correct misspelled words
for word in misspelled_words:
     print(spell_data.correction(word))
# Tokenization function      
def word_tokenizer(text):
    text = text.lower()
    text = text.split()
    return text
import nltk
# Download NLTK stopwords
nltk.download('stopwords')
from nltk.corpus import stopwords
en_stopwords = set(stopwords.words('english'))
print(f"Stop Words in English : \n{ en_stopwords}")
# Removing stopwords from data-text
def remove_stopwords(text):
    text = [word for word in text if word not in en_stopwords]
    return text
from nltk.stem.porter import PorterStemmer
from nltk.stem.lancaster import LancasterStemmer
# Stemming function
stemmer = PorterStemmer()
def stemming(text):
    text = [stemmer.stem(word) for word in text]
    return text
import spacy
sp = spacy.load("en_core_web_sm")

# Define lemmatization function using spacy
def lemmatization(text):
    text = " ".join(text)
    token = sp(text)
    text = [word.lemma_ for word in token]
    return text

# Install langdetect library
!pip install langdetect
from langdetect import detect

# Create a copy of the merged data
merged_data_temp = merged_data.copy()
# Define a function to label sentiments
def label(y):
    if y == '5':
        return 1
    elif y == '4':
        return 1
    else:
        return 0
from tqdm import tqdm
tqdm.pandas()

# Apply the sentiment label function to create a new 'SentimentDecision' column
merged_data_temp['SentimentDecision'] = merged_data_temp.Ratings.progress_map(label)
merged_data_temp = merged_data_temp[['Review', 'SentimentDecision', 'Ratings']]
print(merged_data_temp.head())
merged_data_temp.SentimentDecision.value_counts()
import seaborn as sns
sns.countplot(x = "SentimentDecision", data = merged_data_temp, palette={0: 'red', 1: 'green'})
plt.show()
selected_data = merged_data_temp[(merged_data_temp['Ratings'] == '5')]
positive_reviews = list(selected_data[(selected_data['Review'].str.len() > 100) & (selected_data['Review'].str.len() < 350)]['Review'])
len(positive_reviews)
!pip install nlpaug
import nlpaug.augmenter.word as naw
positive_reviews[41:49]
augmenter = naw.AntonymAug(name='Antonym_Aug', aug_min=1, aug_max=10, aug_p=0.3, lang='eng', stopwords=en_stopwords, tokenizer=None,
                     reverse_tokenizer=None, stopwords_regex=None, verbose=0)
augmented_negative = augmenter.augment(positive_reviews)
len(augmented_negative)
augmented_negative[41:49]
merged_data_negative = pd.DataFrame({"Review" : augmented_negative, 'SentimentDecision' : [0]*len(augmented_negative)})
merged_data_positive = pd.DataFrame({"Review" : positive_reviews, 'SentimentDecision' : [1]*len(positive_reviews)})
merged_data_temp = pd.concat([merged_data_negative, merged_data_positive]).sample(frac = 1, random_state = 11).reset_index(drop=True)
print(merged_data_temp.info())
print(merged_data_temp.head())
merged_data = merged_data_temp
import unicodedata as uni
def preprocessing(text):

    text = remove_url(text)
    text = uni.normalize('NFKD', text)
    text = handle_emoji(text)
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    text = word_tokenizer(text)
    # text = stemming(text)
    text = lemmatization(text)
    text = remove_stopwords(text)
    text = " ".join(text)
    return text
from tqdm import tqdm
tqdm.pandas()
merged_data['Clear_Feedback'] = merged_data['Review'].progress_map(preprocessing)
print(merged_data.head())
reviews = merged_data.Clear_Feedback.values.tolist()
from tqdm import tqdm
tqdm.pandas()
merged_data['Clear_Feedback2'] = merged_data['Clear_Feedback'].progress_map(word_tokenizer)
print(merged_data.head())
data_words = merged_data['Clear_Feedback2'].values.tolist()
len(data_words)
import gensim.corpora as corpora

# Create Dictionary
id2word = corpora.Dictionary(data_words)
# Create Corpus
texts = data_words
# Term Document Frequency
corpus = [id2word.doc2bow(text) for text in texts]
# View
print(corpus[:1][0][:30])
from gensim.models import LdaMulticore
from gensim.models import LdaModel
from pprint import pprint

# number of topics
num_topics = 10
# Build LDA model
lda_model = LdaMulticore(corpus=corpus, id2word=id2word,
                     num_topics=num_topics, iterations=400)
# Print the Keyword in the 10 topics
pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]
from gensim.models import FastText
fasttext_model = FastText(data_words, vector_size= 100, window=5, min_count=5, workers=4,sg=1)
fasttext_model.save("FastText-Model-For-ABSA.bin")
fasttext_model.wv.n_similarity("I truly enjoy the battery of this mobile", "battery")
fasttext_model.wv.n_similarity("I truly enjoy the camera of this mobile", "camera")
aspects = ["mobile", "camera", "battery", "delivery", "processor"]
def get_similarity(text, aspect):
    try:
        text = " ".join(text)
        return fasttext_model.wv.n_similarity(text, aspect)
    except:
        return 0
from tqdm import tqdm
tqdm.pandas()
for aspect in aspects:
    merged_data[aspect] = merged_data['Clear_Feedback2'].progress_map(lambda text: get_similarity(text, aspect))
print(merged_data.head())
merged_data.to_csv("Cleared_Amazon_Products.csv", index = False)
import torch
from torch import nn
from torch.utils.data import Dataset
from torch.utils.data import DataLoader
from torch.utils.data import RandomSampler
import warnings
class config:
    warnings.filterwarnings("ignore", category = UserWarning)
    IMG_SIZE = (224,224)
    DEVICE_TYPE = ("cuda" if torch.cuda.is_available() else "cpu")
    NUM_FOLDS = 5
    SHOULD_SHUFFLE = True
    BATCH_SIZE = 32
    LEARNING_RATE = 0.01
    NUM_EPOCHS = 30
    EMBEDDING_DIMENSION = 100
    MAX_SEQUENCE_LENGTH = 20
    MODEL_FILE_PATH = "TrainedModel.pt"
print(merged_data.head())
class Vocabulary:

    def __init__(self, frequency_threshold, max_size):

        #Establish the mapping for tokens and their corresponding indices.
        #<PAD>: Represents padding for aligning sentence lengths within a batch. 
        #<SOS>: Indicates the initiation of a sentence. 
        #<EOS>: Denotes the conclusion of a sentence. 
        #<UNK>: Substitutes words not present in the vocabulary.
        
        self.itos = {0: '<PAD>', 1:'<SOS>', 2:'<EOS>', 3: '<UNK>'}
        #dictionary for mapping tokens to their respective indices.
        self.stoi = {k:j for j,k in self.itos.items()}

        self.frequency_threshold = frequency_threshold
        self.max_size = max_size

    def __len__(self):
        return len(self.itos)

    @staticmethod
    def tokenizer(text):
        return [tok.lower().strip() for tok in text.split(' ')]


    def build_vocabulary(self, sentence_list):
        #calculate the frequency_occurances of each word first to remove the words with freq < frequency_threshold
        frequency_occurances = {}  #init the freq dictionary
        idx = 4 #index from which we want our dict to start. We already used 4 indexes for pad, start, end, unk

        #calculate freq of words
        for sentence in sentence_list:
            for word in self.tokenizer(sentence):
                if word not in frequency_occurances.keys():
                    frequency_occurances[word]=1
                else:
                    frequency_occurances[word]+=1
        #limit vocab by removing low freq words
        frequency_occurances = {k:v for k,v in frequency_occurances.items() if v>self.frequency_threshold} #limit vocab to the max_size specified
        frequency_occurances = dict(sorted(frequency_occurances.items(), key = lambda x: -x[1])[:self.max_size-idx]) # idx =4 for pad, start, end , unk
        #create vocab
        for word in frequency_occurances.keys():
            self.stoi[word] = idx
            self.itos[idx] = word
            idx+=1

    def numericalize(self, text):
        #tokenize text
        tokenized_text = self.tokenizer(text)
        numericalized_text = []
        for token in tokenized_text:
            if token in self.stoi.keys():
                numericalized_text.append(self.stoi[token])
            else: #out-of-vocab words are represented by UNK token index
                numericalized_text.append(self.stoi['<UNK>'])
        return numericalized_text
from torch.utils.data import Dataset
class CustomDataset(Dataset):

    def __init__(self, merged_data, source_column,frequency_threshold = 3,
                source_vocab_max_size = 10000 , transform=None):
        self.merged_data = merged_data
        self.transform = transform
        #get source and target texts
        self.source_texts = self.merged_data[source_column]
        ##VOCAB class has been created above
        #Initialize source vocab object and build vocabulary
        self.source_vocab = Vocabulary(frequency_threshold, source_vocab_max_size)
        self.source_vocab.build_vocabulary(self.source_texts.tolist())
    def __len__(self):
        return len(self.merged_data)

    def __getitem__(self, index):
        source_text = self.source_texts[index]
        if self.transform is not None:
            source_text = self.transform(source_text)
        #numericalize texts ['<SOS>','cat', 'in', 'a', 'bag','<EOS>'] -> [1,12,2,9,24,2]
        numerialized_source = [self.source_vocab.stoi["<SOS>"]]
        numerialized_source += self.source_vocab.numericalize(source_text)
        numerialized_source.append(self.source_vocab.stoi["<EOS>"])
        #convert the list to tensor and return
        return torch.tensor(numerialized_source), torch.tensor(self.merged_data.SentimentDecision[index])
print(merged_data.head())
dataset = CustomDataset(merged_data, "Clear_Feedback")
len(dataset.source_vocab.stoi)
import pickle

with open('dataset-new', 'wb') as dataset_file:

  # Step 3
    pickle.dump(dataset, dataset_file, pickle.HIGHEST_PROTOCOL)

def get_emb_layer_with_weights(target_vocab, emb_model, trainable = False):

    weights_matrix = np.zeros((len(target_vocab), config.EMBEDDING_DIMENSION))
    words_found = 0

    for i, word in enumerate(target_vocab):
        weights_matrix[i] = np.concatenate([emb_model.wv[word]])
        words_found += 1

    print(f"Words found are : {words_found}")

    weights_matrix = torch.tensor(weights_matrix, dtype = torch.float32).reshape(len(target_vocab), config.EMBEDDING_DIMENSION)
    emb_layer = nn.Embedding.from_pretrained(weights_matrix)
    print(emb_layer)
    if trainable:
        emb_layer.weight.requires_grad = True
    else:
        emb_layer.weight.requires_grad = False

    return emb_layer
class MyCollate:
    def __init__(self, pad_idx, maxlen):
        self.pad_idx = pad_idx
        self.maxlen = maxlen

    def __call__(self, batch):
        #get all source indexed sentences of the batch
        source = [item[0] for item in batch]
        #pad them using pad_sequence method from pytorch.

        padded_sequence = torch.zeros((self.maxlen, len(batch)), dtype = torch.int)

        for idx, text in enumerate(source):

            if len(text) > self.maxlen:
                padded_sequence[:, idx] = source[idx][: self.maxlen]
            else:
                padded_sequence[:len(source[idx]), idx] = padded_sequence[:len(source[idx]), idx] + source[idx]
        #get all target indexed sentences of the batch
        target = [item[1] for item in batch]

        target = torch.tensor(target, dtype = torch.float32).reshape(-1)
        return padded_sequence, target
class Model(nn.Module):
    def __init__(self, input_dim, embedding_dim, hidden_dim, output_dim, embedding_layer):
        super().__init__()
        self.hidden_dim = hidden_dim
        self.embedding = embedding_layer
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, bidirectional = True)
        self.fc1 = nn.Linear(2*hidden_dim, 128)
        self.fc2 = nn.Linear(128, output_dim)
        self.dropout = nn.Dropout(0.3)
        self.sigmoid = nn.Sigmoid()

    def forward(self, text):

        max_len, N = text.shape
        hidden = torch.zeros((2, N , self.hidden_dim),
                          dtype=torch.float)
        memory = torch.zeros((2, N , self.hidden_dim),
                          dtype=torch.float)
        hidden = hidden.to(config.DEVICE_TYPE)
        memory = memory.to(config.DEVICE_TYPE)
        embedded = self.embedding(text)
        output, hidden = self.lstm(embedded, (hidden, memory))
        y_pred = output[-1,:,:]
        y_pred = self.fc1(y_pred)
        y_pred = self.fc2(y_pred)
        y_pred = self.sigmoid(y_pred)
        return y_pred
def train_epochs(dataloader,model, loss_fn, optimizer):
    train_correct = 0
    train_loss = 0

    model.train()

    for Review, label in tqdm(dataloader):

        Review, label = Review.to(config.DEVICE_TYPE), label.to(config.DEVICE_TYPE)
        optimizer.zero_grad()
        output = model(Review)
        output = output.reshape(-1)
        loss = loss_fn(output, label)
        loss.backward()
        optimizer.step()

        train_loss += loss.item()*Review.size(1)
        prediction = (output > 0.5).float()
        train_correct += (prediction == label).float().sum()

    return train_loss, train_correct
def val_epochs(dataloader, model, loss_fn):
    val_correct = 0
    val_loss = 0

    model.eval()

    for Review, label in dataloader:

        Review, label = Review.to(config.DEVICE_TYPE), label.to(config.DEVICE_TYPE)

        output = model(Review)
        output = output.reshape(-1)

        loss = loss_fn(output, label)

        val_loss += loss.item()*Review.size(1)
        prediction = (output > 0.5).float()
        val_correct += (prediction == label).float().sum()
    return val_loss, val_correct
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import KFold
from torch.utils.data import SubsetRandomSampler
from torch.optim import Adam
from tqdm import tqdm
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt

kfold = KFold(n_splits = config.NUM_FOLDS)
model_state_dicts = {}

for fold, (train_idx, val_idx) in enumerate(kfold.split(np.arange(len(dataset)))):

    train_sampler = SubsetRandomSampler(train_idx)
    val_sampler = SubsetRandomSampler(val_idx)

    train_loader = DataLoader(dataset, batch_size = config.BATCH_SIZE, sampler = train_sampler, collate_fn = MyCollate(0, config.MAX_SEQUENCE_LENGTH))
    val_loader = DataLoader(dataset, batch_size = config.BATCH_SIZE, sampler = val_sampler, collate_fn = MyCollate(0, config.MAX_SEQUENCE_LENGTH))

    VOCAB_SIZE = len(dataset.source_vocab)
    HIDDEN_DIM = 128
    OUTPUT_DIM = 1
    VOCAB = list(dataset.source_vocab.stoi)
    embedding_layer = get_emb_layer_with_weights(target_vocab = VOCAB, emb_model = fasttext_model, trainable = False)

    model = Model(VOCAB_SIZE, config.EMBEDDING_DIMENSION, HIDDEN_DIM, OUTPUT_DIM, embedding_layer)
    model = model.to(config.DEVICE_TYPE)

    loss_fn = nn.BCELoss()
    optimizer = torch.optim.SGD(model.parameters(), lr = 0.1)

    train_losses = []
    val_losses = []
    train_accs = []
    val_accs = []

    print(f"****************************{fold}-Model Fold****************************")
    for epoch in range(config.NUM_EPOCHS):
        train_loss, train_correct = train_epochs(train_loader, model, loss_fn, optimizer)
        val_loss, val_correct = val_epochs(val_loader, model, loss_fn)

        train_loss = train_loss/len(train_loader.sampler)
        val_loss = val_loss/len(val_loader.sampler)
        train_acc = (train_correct/len(train_loader.sampler))*100
        val_acc = (val_correct/len(val_loader.sampler))*100

        train_losses.append(train_loss)
        val_losses.append(val_loss)
        train_accs.append(train_acc.cpu().numpy().tolist())
        val_accs.append(val_acc.cpu().numpy().tolist())

        print(f"| Train Loss : {train_loss} |", end = " ")
        print(f" Val Loss : {val_loss} |", end = " ")
        print(f"Train Acc : {train_acc} |", end = " ")
        print(f"Val Acc : {val_acc} |")

    # Saving the state dicts for the model
    model_state_dicts.update({f"LSTM-Model-for-{fold}" : model.state_dict(),
                             f"Model-Optimizer-for-{fold}" : optimizer.state_dict()})

plt.plot(train_accs)
plt.plot(val_accs)
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(train_losses)
plt.plot(val_losses)
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

torch.save(model_state_dicts, "My-Model.pt")
def numericalize(text):

    numerialized_source = []
    numerialized_source = [dataset.source_vocab.stoi["<SOS>"]]
    numerialized_source += dataset.source_vocab.numericalize(text)
    numerialized_source.append(dataset.source_vocab.stoi["<EOS>"])

    return numerialized_source

def padding(source):
    padded_sequence = torch.zeros(config.MAX_SEQUENCE_LENGTH, 1, dtype = torch.int)
    source = torch.tensor(source)

    if len(source) > config.MAX_SEQUENCE_LENGTH:
        padded_sequence[:, 0] = source[: config.MAX_SEQUENCE_LENGTH]
    else:
        padded_sequence[:len(source), 0] = padded_sequence[:len(source), 0] + source

    return padded_sequence
def infer_processing(text):

    text = preprocessing(text)
    text = numericalize(text)
    text = padding(text)
    return text
aspects = ["mobile", "camera", "battery", "neutral", "processor"]

def get_similarity(text, aspect):
    try:
        return fasttext_model.wv.n_similarity(text, aspect)
    except:
        return 0

def best_aspect(text, aspects):
    a = []

    for aspect in aspects:
        a.append(get_similarity(text, aspect))

    return aspects[np.argmax(a)]
 
    ########################################
# Function to process the input and print the result
def process_and_print(sample):
    
    # Determine the best aspect from the preprocessed sample
    bestaspect = best_aspect(preprocessing(sample), aspects)
    
    # Infer processing and move data to the specified device (e.g., GPU)
    processed_data = infer_processing(sample).to(config.DEVICE_TYPE)
    model.eval()
    # Get the sentiment prediction from the model
    sentiment = model(processed_data)
    sentiment = sentiment.cpu().detach().numpy()[0]
    # Classify sentiment as optimistic or pessimistic based on the prediction
    if sentiment > 0.5:
        sentiment_label = 'optimistic'
    else:
        sentiment_label = 'pessimistic'
    # Set the result string for display
    result.set(f"The {bestaspect} of the product has drawn {sentiment_label} remarks from the Reviewer in their feedback")
    # Print the result to the console
    print(f"The {bestaspect} of the product has drawn {sentiment_label} remarks from the Reviewer in their feedback")
# Function to create the GUI and return the entries
def make_gui(root, fields):
    entries = {}
    for field in fields:
        row = Frame(root)
        lab = Label(row, width=14, text=field + ": ", anchor='w')
        ent = Entry(row)
        ent.insert(0, "")
        row.pack(side=TOP, fill=X, padx=5, pady=5)
        lab.pack(side=LEFT)
        ent.pack(side=RIGHT, expand=YES, fill=X)
        entries[field] = ent
    return entries
# Function to process the text entered in the GUI
def process_text(entries):
    global sample
    sample = entries['Sample Input'].get()
    # Check if the input is alphabetic or contains spaces
    if sample.isalpha() or " " in sample:
        process_and_print(sample)
        root.withdraw()  # Withdraw the GUI (hide it)
        root.quit()  # Quit the Tkinter main loop
    else:
        result.set("Invalid input. Please enter text only.")

# Create the main GUI window
root = Tk()
root.title("Text Input GUI")
# Define the fields for the GUI
fields = ('Sample Input',)

# Create the GUI and get the entries
ents = make_gui(root, fields)

# Variable to store the result string
result = StringVar()
result.set("Result: ")
result_label = Label(root, textvariable=result)
result_label.pack(pady=10)
# Button to submit the input and trigger processing
submit_button = Button(root, text='Submit', command=lambda e=ents: process_text(e))
submit_button.pack(pady=10)

sample = ""  # Variable to store the entered text
# Start the Tkinter main loop
root.mainloop()
```
